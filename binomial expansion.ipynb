```python
import numpy as np
from sympy import binomial as bi
from decimal import *
from math import comb
import scipy
```

```python
def symby_binomial_expansion(x,n):
    sumed=[]
    for k in range(0,n+1,2):
        sumed.append(bi(n,k)*Decimal(x**k) + bi(n,k+1)*Decimal(x**(k+1)))
    print("sympy_binomial_expansion:", np.sum(sumed))

def math_binomial_expansion(x,n):
    sumed=[]
    for k in range(0,n+1):
        sumed.append(comb(n,k)*Decimal(x**k))
    print("math_binomial_expansion:",np.sum(sumed))
def my_binomial_expansion(x,n):
    sumed=[]
    for k in range(0,n+1,2):
        sumed.append(comb(n,k)*Decimal(x**k) + comb(n,k+1)*Decimal(x**(k+1)))
    print("my_binomial_expansion:",np.sum(sumed))
```

```python
# for eta in np.linspace(0.05, 0.9, 10):
eta=0.5
for L in range(10,205):
    symby_binomial_expansion(x=-eta,n=L)
    math_binomial_expansion(x=-eta,n=L)
    my_binomial_expansion(x=-eta,n=L)
    print((1-eta)**L)
    print("-"*20)
#     
```

```python
from math import comb
def check_equality2(eta, L):
    
    def compute_upper_bound(eta, L):
        sumed=Decimal(0)
        for k in range(1,2*L+1):
            exponent=Decimal((-eta)**k)
            term=comb(2*L-1,k-1)-comb(2*L-3,k-2)
            sumed+=Decimal(exponent*term)
        print("expanded upper bound:", sumed)
        return sumed
    
    def compute_equality_right(eta, L):
        RHS=-eta*(1-eta)**(2*L-1)  - (eta**2)* ((1-eta)**(2*L-3))
        print("right hand side:", RHS)
        return RHS
    compute_upper_bound(eta, L)
    compute_equality_right(eta, L)
for eta in np.linspace(0.01, 0.9, 10):
    for L in range(3,20):
        check_equality2(eta=eta, L=L)
```
