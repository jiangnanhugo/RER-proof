```python
def compute_LHS(eta, L):
    LHS=0
    for k in range(1,2*L+1):
        exponent=(-eta)**k
        term=0
        for l in range(1, L+1):
            term+=bi(L+l-2, k-1)+ bi(L-l,k-2) #+ bi(2*l-2, k-2)
        LHS+=exponent*term
    print("left hand side:", LHS)
    return LHS

def compute_RHS(eta, L):
    RHS=(1-eta)**(2*L)-(1-eta)**(2*L-2)+(1-eta)**(L)-(1-eta)**(L-2)
    print("right hand side:", RHS)
    return RHS
```

```python
compute_LHS(eta, L)
compute_RHS(eta, L)
```

```python
def binomial_expansion(a,b,n):
    sumed=[]
    for k in range(0,n+1):
        sumed.append(bi(n,k)*a**(n-k)*b**k)
    print("binomial_expansion:")
    print((a+b)**n)
    print(1+np.sum(sumed))
    print(sumed)
```

```python
binomial_expansion(a=1,b=-eta,n=2*L)
binomial_expansion(a=1,b=-eta,n=L)
binomial_expansion(a=1,b=-eta,n=L-2)
binomial_expansion(a=1,b=-eta,n=2*L-3)
```

```python
def check_equality(eta, L):
    def compute_equality_left(eta, L):
        LHS=0
        for k in range(1,2*L+1):
            exponent=(-eta)**k
            for l in range(1, L+1):
                LHS+=exponent *(bi(L+l-2, k-1) + bi(L-l, k-1))
        print("left hand side:", LHS)
        return LHS
    
    def compute_equality_right(eta, L):
        RHS=(1-eta)**(2*L-1)-(1-eta)**(L-1) + (1-eta)**(L)-1
        print("right hand side:", RHS)
        return RHS
    compute_equality_left(eta, L)
    compute_equality_right(eta, L)
check_equality(eta=0.5, L=10)
check_equality(eta=0.4, L=10)
check_equality(eta=0.2, L=20)
```
