```python
def equal(x1, x2,N=14):
    if round(x1, N) == round(x2,N):
        return True
    return False
```

```python
def check_equality(eta, L):
    def compute_equality_left(eta, L):
        LHS=Decimal(0)
        for k in range(1,2*L+1):
            exponent=pow(Decimal(-eta),Decimal(k))
            for l in range(1, L+1):
                LHS+=exponent *(bi(L+l-2, k-1) + bi(L-l, k-1))
        print("LHS:", LHS, end="\t")
        return LHS
    
    def compute_equality_right(eta, L):
        RHS=pow(Decimal(1-eta), Decimal(2*L-1)) \
            - pow(Decimal(1-eta), Decimal(L-1)) \
            + pow(Decimal(1-eta),Decimal(L))\
            - Decimal(1)
        print("RHS:", RHS, end="\t")
        return RHS
    print(equal(compute_equality_left(eta, L),compute_equality_right(eta, L)))
for eta in np.linspace(0.01, 0.9, 100):
    for L in range(10,39):
        check_equality(eta=eta, L=L)
```

```python
def check_equality2(eta, L):
    def compute_equality_left(eta, L):
        LHS=0
        for k in range(2,2*L+1):
            term=[]
            for l in range(1, L+1):
                if 2*l-2 < k-2:
                    continue
                term.append(bi(2*l-2, k-2))
            LHS+=np.sum(term)*Decimal((-eta)**k)
        print("LHS: {}".format(LHS),end="\t")
        return LHS
    
    def compute_upper_bound(eta, L):
        sumed=Decimal(0)
        for k in range(2,2*L+1):
            exponent=Decimal((-eta)**k)
            term=bi(2*L-1,k-1)-bi(2*L-3,k-2)
            sumed+=Decimal(exponent*term)
        print("UB:", sumed, end="\t")
        return sumed
    
    def compute_equality_right(eta, L):
        RHS=Decimal(eta)\
            - Decimal(eta)*pow(Decimal(1-eta), Decimal(2*L-1)) \
            - Decimal(eta) * Decimal(eta) * pow(Decimal(1-eta), Decimal(2*L-3)) 
#         RHS=eta - eta*(1-eta)**(2*L-1)  - (eta**2)* ((1-eta)**(2*L-3))

        print("RHS:", RHS)
        return RHS
    print(compute_equality_left(eta, L)< compute_upper_bound(eta, L))
    compute_equality_right(eta, L)
for eta in np.linspace(0.01, 0.9, 10):
    for L in range(3,20):
        check_equality2(eta=eta, L=L)
```

```python
def compute_bi(n,k):
    ratio=Decimal(1)
    for i in range(k):
        ratio=ratio*Decimal((n-i)/(k-1))
    return ratio
for n in range(100, 120):
    for k in range(50, 60):
        print(compute_bi(n=n,k=k),bi(n,k))
```
